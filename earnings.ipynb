### Exploring the data

* Let's setup the environment by importing the libraries we need
* Read the dataset into a DataFrame and start exploring the data
* Drop rows with missing values

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")

print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())

recent_grads.describe()

raw_data_count = (recent_grads.shape)[0]
recent_grads = recent_grads.dropna()
cleaned_data_count = (recent_grads.shape)[0]

print("Rows before dropna:",raw_data_count, "\nRows after dropna:", cleaned_data_count)
```

### Pandas Histograms

Exploring Data using Visualizations: histograms

```python
cols = ["Sample_size","Median","Employed","Full_time","ShareWomen","Unemployment_rate","Men","Women"]

fig = plt.figure(figsize=(10,20))

for i in range(0,8):
    ax = fig.add_subplot(4,2,i+1)
    max_val = recent_grads[cols[i]].max()
    ax = recent_grads[cols[i]].hist(xrot=40, bins=20, range=(0,max_val))
    ax.set_ylabel("Frequency")
```
